## Exploratory Data Analysis

```python
plt.plot(monthly_avg.index, monthly_avg['Temperature'], marker='o')
plt.title("Average Monthly Temperature")
plt.xlabel("Month")
plt.ylabel("Temperature")
plt.grid()
plt.show()
```

### 📊 Insight

### Temperature fluctuates over time without a consistent upward or downward trend, suggesting changing environmental conditions rather than a steady pattern.

“Temperature fluctuates over time without a consistent upward trend, suggesting changing weather conditions rather than steady growth.”

🌡️ Temperature

* Temperature ranges from 19°C to 46°C, showing significant variation
* Average temperature is around 31°C, indicating generally warm conditions
* There are noticeable fluctuations, suggesting possible seasonal effects





```python
plt.bar(monthly_precip.index, monthly_precip)
plt.title("Monthly Precipitation")
plt.xlabel("Month")
plt.ylabel("Precipitation")
plt.show()
```

```python
monthly_avg['Temperature'].plot(kind='bar', edgecolor='black')

plt.title("Monthly Average Temperature")
plt.xlabel("Month")
plt.ylabel("Temperature")
plt.show()
```

### 📊 Insight

Certain months show higher average temperatures, while others are cooler, indicating seasonal variation. Rainfall is concentrated in specific months, suggesting uneven precipitation patterns.

###  📊 Insight
- Month X shows the highest average temperature
- Rainfall is concentrated in specific months
- Indicates seasonal variation

📅 Monthly Trends




*   Some months show higher average temperature than others
*   Rainfall is concentrated in specific months

*   Indicates seasonal weather patterns





```python
high_temp = np.mean(df['Temperature']) + 2*np.std(df['Temperature'])
heavy_rain = np.mean(df['Precipitation']) + 2*np.std(df['Precipitation'])

extreme_temp = df[df['Temperature'] > high_temp]
extreme_rain = df[df['Precipitation'] > heavy_rain]

print("Extreme Temperature Days:\n", extreme_temp)
print("Extreme Rainfall Days:\n", extreme_rain)
```

The distribution, trend, and correlation analysis together indicate that weather conditions are variable but not strongly interdependent, suggesting that external seasonal factors may play a role.
